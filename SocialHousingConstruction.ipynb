```python
from pymongo import MongoClient
client = MongoClient("mongodb://%s:%s@127.0.0.1" % ("dap", "dap"))
```

```python
db = client['DAPGRPM_database']
```

```python
#Set Collection
Collection = db["social_housing_construction"]
```

```python
import json
# Loading or Opening the json file
with open('social_housing_construction_status_report_q1_2021.json',encoding='latin-1') as file:
    file_data = json.load(file)
     
```

```python
# Inserting the loaded data in the Collection
# if JSON contains data more than one entry
# insert_many is used else insert_one is used
if isinstance(file_data, list):
    Collection.insert_many(file_data) 
else:
    Collection.insert_one(file_data)
```

```python
for x in Collection.find():
    print(x)
```
